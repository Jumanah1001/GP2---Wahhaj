```python
#import library
from AhpAnpLib import inputs_AHPLib as input
from AhpAnpLib import structs_AHPLib as str
from AhpAnpLib import calcs_AHPLib as calc
from AhpAnpLib import ratings_AHPLib as rate
```

```python
#create model
locationsModel=str.Model("AHP Location Selection")

#create goal cluster
cluster0=str.Cluster("1Goal",0)
#create goal node
goal_node=str.Node("GoalNode",0)
#add goal ndoe to the goal cluster
cluster0.addNode2Cluster(goal_node)
#add goal cluster to the model
locationsModel.addCluster2Model(cluster0)

#create criteria cluster and nodes
solar_radiance=str.Node("1solar radiance",1)
sunshine_hours=str.Node("2sunshine hours",2)
slope=str.Node("3slope",3)
elevation=str.Node("4elevation",4)
surface_temprature=str.Node("5surface temprature",5)
cluster1=str.Cluster("2Criteria",1)
cluster1.addMultipleNodes2Cluster(solar_radiance,sunshine_hours,slope,elevation,surface_temprature)
locationsModel.addCluster2Model(cluster1)



# ===============================
# ALTERNATIVES CLUSTER
# ===============================

# Create alternative nodes (temporary placeholders if no real areas yet)
area1 = str.Node("Candidate Area 1", 1)
area2 = str.Node("Candidate Area 2", 2)
area3 = str.Node("Candidate Area 3", 3)

# Create alternatives cluster
alternativesCluster = str.Cluster("3Alternatives", 3)

# Add alternatives to cluster
alternativesCluster.addMultipleNodes2Cluster(area1, area2, area3)

# Add alternatives cluster to model
locationsModel.addCluster2Model(alternativesCluster)
```

```python
#add connections
locationsModel.addNodeConnectionFromAllNodesToAllNodesOfCluster("GoalNode","2Criteria")
locationsModel.addNodeConnectionFromAllNodesToAllNodesOfCluster("2Criteria","3Alternatives")

locationsModel.remNodeConnectionFromTo("1solar radiance","Candidate Area 1")
locationsModel.remNodeConnectionFromTo("1solar radiance","Candidate Area 2")
locationsModel.remNodeConnectionFromTo("1solar radiance","Candidate Area 3")

locationsModel.remNodeConnectionFromTo("2sunshine hours","Candidate Area 1")
locationsModel.remNodeConnectionFromTo("2sunshine hours","Candidate Area 2")
locationsModel.remNodeConnectionFromTo("2sunshine hours","Candidate Area 3")

locationsModel.remNodeConnectionFromTo("3slope","Candidate Area 1")
locationsModel.remNodeConnectionFromTo("3slope","Candidate Area 2")
locationsModel.remNodeConnectionFromTo("3slope","Candidate Area 3")

locationsModel.remNodeConnectionFromTo("4elevation","Candidate Area 1")
locationsModel.remNodeConnectionFromTo("4elevation","Candidate Area 2")
locationsModel.remNodeConnectionFromTo("4elevation","Candidate Area 3")

locationsModel.remNodeConnectionFromTo("5surface temprature","Candidate Area 1")
locationsModel.remNodeConnectionFromTo("5surface temprature","Candidate Area 2")
locationsModel.remNodeConnectionFromTo("5surface temprature","Candidate Area 3")

locationsModel.printStruct()
```

```python
# =====================================
# ADD HIERARCHICAL CONNECTIONS
# =====================================

# Connect Goal cluster to Criteria cluster
locationsModel.addNodeConnectionFromAllNodesToAllNodesOfCluster("1Goal", "2Criteria")

# Connect Criteria cluster to Alternatives cluster
locationsModel.addNodeConnectionFromAllNodesToAllNodesOfCluster("2Criteria", "3Alternatives")

locationsModel.printStruct()
```
